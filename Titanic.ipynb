```python
train["EmbarkedNum"]=train["Embarked"].map({"S":0,"Q":1,"C":2})
train["SexNum"]=train["Sex"].map({"male":0,"female":1})
test["EmbarkedNum"]=test["Embarked"].map({"S":0,"Q":1,"C":2})
test["SexNum"]=test["Sex"].map({"male":0,"female":1})
```

```python
train=train.drop("Ticket",axis=1)
test=test.drop("Ticket",axis=1)
```

```python
train["TitleNum"]=train["Title"].map({"Mr":0,"Miss":1,"Mrs":2,"Master":3,"Dr":4,"Rev":4,"Mlle":4,"Major":4,"Col":4,"Countess":4,"Mme":4,"Lady":4,"Jonkheer":4,"Sir":4,"Capt":4,"Don":4,"Ms":4})

test["TitleNum"]=test["Title"].map({"Mr":0,"Miss":1,"Mrs":2,"Master":3,"Rev":4,"Col":4,"Ms":4,"Dona":4,"Dr":4})
```

```python
train=train.drop(["Name","Sex","Title","Embarked"],axis=1)
test=test.drop(["Name","Sex","Title","Embarked"],axis=1)
```

```python
train=sklearn.utils.shuffle(train)
test=sklearn.utils.shuffle(test)
```

```python
train.head()
```

```python
X=train.drop("Survived",axis = 1)
y=train["Survived"]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split (X,y,test_size=0.2)
```

**MULTINOMIAL NAIVE BAYES**

```python
from sklearn.naive_bayes import MultinomialNB
MNB = MultinomialNB()
MNB.fit(X_train,y_train)
```

```python
y_pred = MNB.predict(X_test)
MNB_acc = metrics.accuracy_score(y_test,y_pred)
print("Accuracy of MultinomialNB is ",MNB_acc*100)
```

**BERNOULLI NAIVE BAYES**

```python
from sklearn.naive_bayes import BernoulliNB
BNB= BernoulliNB()
BNB.fit(X_train,y_train)
```

```python
y_pred = BNB.predict(X_test)
BNB_acc = metrics.accuracy_score(y_test,y_pred)
print("Accuracy of BernoulliNB is",BNB_acc*100)
```

**LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
LR= LogisticRegression(max_iter=1000).fit(X_train,y_train)
```

```python
y_pred = LR.predict(X_test)
LR_acc = metrics.accuracy_score(y_test,y_pred)
print("Accuracy of LR is",LR_acc*100)
```

**SGD CLASSIFIER**

```python
from sklearn.linear_model import SGDClassifier
SGD= SGDClassifier()
SGD.fit(X_train,y_train)
```

```python
y_pred = SGD.predict(X_test)
SGD_acc = metrics.accuracy_score(y_test,y_pred)
print("Accuracy of SGD is",SGD_acc*100)
```

**LINEAR SVC**

```python
from sklearn.svm import LinearSVC
LSVC_Model = LinearSVC()
LSVC_Model.fit(X_train,y_train)
```

```python
y_pred=LSVC_Model.predict(X_test)
LSVCAcc=metrics.accuracy_score(y_test,y_pred)
print('Acuuracy of Linear SVC Model is =',LSVCAcc*100)
```

**RANDOM FOREST**

```python
from sklearn.ensemble import RandomForestClassifier
RFC_Model = RandomForestClassifier()
RFC_Model.fit(X_train,y_train)
```

```python
y_pred=RFC_Model.predict(X_test)
RFCAcc=metrics.accuracy_score(y_test,y_pred)
print('Acuuracy of RandomForestClassifier Model is =',RFCAcc*100)
```
